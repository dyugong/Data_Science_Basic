```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests

try:
    huck_finn_url = 'https://www.inferentialthinking.com/data/huck_finn.txt'
    huck_finn_text = requests.get(huck_finn_url)
    #print("HTML:\n", huck_finn_text.text)
except:
    print("Invalid huck_finn_url or some error occured while making the GET request to the specified huck_finn_url")

huck_finn_chapters = huck_finn_text.text.split('CHAPTER ')[44:]

try:
    little_women_url = 'https://www.inferentialthinking.com/data/little_women.txt'
    little_women_text = requests.get(little_women_url)
    #print("HTML:\n", huck_finn_text.text)
except:
    print("Invalid little_women_url or some error occured while making the GET request to the specified little_women_url")

little_women_chapters = little_women_text.text.split('CHAPTER ')[1:]
```

```python
# Display the chapters of Huckleberry Finn in a table.
#Table().with_column('Chapters', huck_finn_chapters)

df = pd.DataFrame({'Chapters': huck_finn_chapters})
print(df)
```

```python
# Count how many times the names Jim, Tom, and Huck appear in each chapter.
counts = pd.DataFrame({'Jim': np.char.count(huck_finn_chapters, 'Jim'), 'Tom': np.char.count(huck_finn_chapters, 'Tom'), 'Huck': np.char.count(huck_finn_chapters, 'Huck')})

counts
```

```python
# Plot the cumulative counts:
# how many times in Chapter 1, how many times in Chapters 1 and 2, and so on.
cum_counts = counts.cumsum()

cum_counts
```

```python
cum_counts.plot.line(figsize = (10,5))
```

```python
# The chapters of Little Women, in a table

#Table().with_column('Chapters', little_women_chapters)

lwc = pd.DataFrame({'Chapters': little_women_chapters})
print(lwc)
```

```python
# Counts of names in the chapters of Little Women

lwc_counts = pd.DataFrame({'Amy': np.char.count(little_women_chapters, 'Amy'), 'Beth': np.char.count(little_women_chapters, 'Beth'), 'Jo': np.char.count(little_women_chapters, 'Jo'), 'Meg': np.char.count(little_women_chapters, 'Meg'), 'Laurie': np.char.count(little_women_chapters, 'Laurie')})

lwc_counts
```

```python
# Plot the cumulative counts

lwc_cum_counts = lwc_counts.cumsum()
lwc_cum_counts
```

```python
lwc_cum_counts.plot.line(figsize = (10,5))
```

```python
# In each chapter, count the number of all characters
# call this the "length" of the chapter
# Also count the number of periods

# Huck Finn
chars_periods_huck_finn = pd.DataFrame({'Huck Finn Chapter Length':np.char.str_len(huck_finn_chapters), 'Number of Periods': np.char.count(huck_finn_chapters,'.')})
chars_periods_huck_finn
```

```python
# Little Women

chars_periods_little_women = pd.DataFrame({'Little Women Chapter Length':np.char.str_len(little_women_chapters), 'Number of Periods': np.char.count(little_women_chapters,'.')})
chars_periods_little_women
```

```python
chars_periods_huck_finn[['Number of Periods']]
```

```python
import matplotlib.pyplot as plt

huck_period = chars_periods_huck_finn['Number of Periods']
huck_length = chars_periods_huck_finn['Huck Finn Chapter Length']

little_period = chars_periods_little_women['Number of Periods']
little_length = chars_periods_little_women['Little Women Chapter Length']

plt.scatter(x=huck_period, y=huck_length, s=10, c='royalblue')
plt.scatter(x=little_period, y=little_length, s=10, c='olive')

plt.xlabel('Number of periods in chapter')
plt.ylabel('Number of characters in chapter')

plt.show()
```
